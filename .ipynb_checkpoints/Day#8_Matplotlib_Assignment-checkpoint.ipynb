```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
data = {'unemployment_rate': [6.1, 5.8, 5.7, 5.7, 5.8, 5.6, 5.5, 5.3, 5.2, 5.2],
        'index_price': [1500, 1520, 1525, 1523, 1515, 1540, 1545, 1560, 1555, 1565]
       }
  
df = pd.DataFrame(data)
print(df)
```

```python
plt.plot(data['index_price'],data['unemployment_rate'], c = '0.4')

plt.xlabel('index_price')
plt.ylabel('unemployment_rate')
plt.title('Rate of unemployment')

plt.xlim(1500,1565)

plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Tomatoes", "Mangoes", "Oranges", "Apples"]
myexplode = [0, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

## Q5: Create three line charts with Legends at left and colors

## Q6: Create subplots with three charts and figsize=20,10

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 10))

x = np.arange(0, 5, 0.25)

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("First")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("Second ")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("Third");
```
